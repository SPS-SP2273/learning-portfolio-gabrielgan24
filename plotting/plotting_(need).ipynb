```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')  #added
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,  #ERROR BARS ADDED
             color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.1)
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
#ANOTHER Y AXIS with same x axis => twinx()

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

```python
#too few points, to fill up use np.linspace

x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')


plt.show()
```

### 1.8 Saving to disc

## Exercise 2 :  A simple mathematical plot

```python
#DATA POINTS

x = [0]
for i in range(1,51):
    x.append(i)

#PLOTTING

x = np.linspace(0,50,300)

sin_x = np.sin(x)
negative = np.negative(x)
e_power = np.exp(negative/10)
y = (sin_x)*e_power

plt.plot(x,y)
plt.ylim(-1,1)  #4
plt.grid(alpha=0.25)  #2
plt.xlabel('x', fontsize=15, color='darkblue')
plt.ylabel('y', fontsize=15, color='darkblue')  #1
plt.title('$y=e^{-x/10}/sinx(x)$', fontsize=20)  #3
plt.show()
```
